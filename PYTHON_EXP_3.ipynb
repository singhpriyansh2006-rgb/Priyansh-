```python
#WAP to Find a factorial of given number
num = int(input("Enter a number: "))
factorial = 1

if num < 0:
    print("Factorial does not exist for negative numbers")
elif num == 0:
    print("Factorial of 0 is 1")
else:
    for i in range(1, num + 1):
        factorial = factorial * i
    print("Factorial of", num, "is", factorial)
```

```python
#WAP to Find whether the given number is Armstrong numbe
num = int(input("Enter a number: "))

temp = num
order = len(str(num))
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** order
    temp //= 10
if sum == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is NOT an Armstrong number")
```

```python
#WAP to  Print Fibonacci series up to given term.
n = int(input("Enter number of terms: "))

a = 0
b = 1

print("Fibonacci Series:")

if n <= 0:
    print("Please enter a positive number")
elif n == 1:
    print(a)
else:
    print(a)
    print(b)
    for i in range(2, n):
        c = a + b
        print(c)
        a = b
        b = c
```

```python
#WAP toWrite a program to find if given number is prime number or not
num = int(input("Enter a number: "))

if num <= 1:
    print(num, "is NOT a Prime Number")
else:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is NOT a Prime Number")
            break
    else:
        print(num, "is a Prime Number")
```

```python
#WAP to Check whether given number is palindrome or not.
num = int(input("Enter a number:"))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10
if original == reverse:
    print(original, "is a Palindrome number")
else:
    print(original, "is NOT a Palindrome number")
```

```python
#WAP to Write a program to print sum of digits.
num = int(input("Enter a number:"))
temp = num
sum_of_digits = 0

while temp > 0:
    digit = temp % 10
    sum_of_digits += digit
    temp //= 10
print("Sum of digits of", num, "is", sum_of_digits)
```

```python
#WAP to Count and print all numbers divisible by 5 or 7 between 1 to 100
count = 0

print("Numbers divisible by 5 or 7 between 1 and 100:")

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("Total count =", count)
```

```python
#WAP to Write a program to print the following pattern
n = 5

for i in range(n, 0, -1):
    for j in range(1, i + 1):
        print(j, end="")
    for k in range(n - i):
        print("*", end="")
    for k in range(n - i):
        print("*", end="")
    for j in range(i, 0, -1):
        print(j, end="")

    print()
```
