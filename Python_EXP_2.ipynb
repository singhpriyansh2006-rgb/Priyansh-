```python
#Check whether the quadratic equation has real roots or imaginary roots. Display the root


import math
a = float(input("Enter value of a:"))
b = float(input("Enter value of b:"))
c = float(input("Enter value of c:"))

D = b**2 - 4*a*c

# Check nature of roots
if D > 0:
    root1 = (-b + math.sqrt(D)) / (2*a)
    root2 = (-b - math.sqrt(D)) / (2*a)
    print("Roots are real and distinct")
    print("Root 1 =", root1)
    print("Root 2 =", root2)

elif D == 0:
    root = -b / (2*a)
    print("Roots are real and equal")
    print("Root =", root)

else:
    real_part = -b / (2*a)
    imaginary_part = math.sqrt(-D) / (2*a)
    print("Roots are imaginary")
    print("Root 1 =", real_part, "+", imaginary_part, "i")
    print("Root 2 =", real_part, "-", imaginary_part, "i")
```

```python
#WAP Find whether a given year is a leap year or not.
year = int(input("Enter a year:"))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a Leap Year")
else:
    print(year, "is NOT a Leap Year")
```

```python
#WAP to Write a program which takes any date as input and display next date of the calenda

day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

days_in_month = [31, 28, 31, 30, 31, 30,
                 31, 31, 30, 31, 30, 31]

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    days_in_month[1] = 29

day += 1

if day > days_in_month[month - 1]:
    day = 1
    month += 1
    if month > 12:
        month = 1
        year += 1
print("Next date is:")
print("Day =", day, "Month =", month, "Year =", year)
```

```python
#WAP to Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and
#calculate the percentage
name = input("Enter Name: ")
roll = input("Enter Roll Number: ")
sapid = input("Enter SAPID: ")
sem = input("Enter Semester: ")
course = input("Enter Course: ")

print("\nEnter marks of 5 subjects:")
sub1 = float(input("PDS: "))
sub2 = float(input("Python: "))
sub3 = float(input("Chemistry: "))
sub4 = float(input("English: "))
sub5 = float(input("Physics: "))

total = sub1 + sub2 + sub3 + sub4 + sub5
percentage = total / 5

cgpa = percentage / 10

if 0 <= cgpa <= 3.4:
    grade = "F"
elif 3.5 <= cgpa <= 5.0:
    grade = "C+"
elif 5.1 <= cgpa <= 6.0:
    grade = "B"
elif 6.1 <= cgpa <= 7.0:
    grade = "B+"
elif 7.1 <= cgpa <= 8.0:
    grade = "A"
elif 8.1 <= cgpa <= 9.0:
    grade = "A+"
elif 9.1 <= cgpa <= 10:
    grade = "O (Outstanding)"
else:
    grade = "Invalid CGPA"

print("\n----- GRADE SHEET -----")
print("Name:", name)
print("Roll Number:", roll, "SAPID:", sapid)
print("Sem:", sem, "Course:", course)

print("\nSubject Name : Marks")
print("PDS:", sub1)
print("Python:", sub2)
print("Chemistry:", sub3)
print("English:", sub4)
print("Physics:", sub5)

print("\nPercentage:", round(percentage, 2), "%")
print("CGPA:", round(cgpa, 2))
print("Grade:", grade)
```
