```python
#1.Write a program to count and display the number of capital letters in a given string.
text = input("Enter a string: ")
count = 0
for ch in text:
    if ch.isupper():
       count += 1
print("Number of capital letters:", count)
```

```python
#2. count total number of vowels in string.
text = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for ch in text:
    if ch in vowels:
       count += 1
print("Total number of vowels:", count)
```

```python
#4. WAP to enter a string and substring. you have to print the number of times that the substring occurs in the given string. string
#traversal will take place from left to right not from right to left.
string = input("Enter the main string: ")
substring = input("Enter the substring: ")

count = 0
i = 0
len_sub = len(substring)
while i <= len(string) - len_sub:
      if string[i:i+len_sub] == substring:
          count += 1
          i += len_sub
else:
    i += 1
print("Substring occurs", count, "times")
```

```python
#5.Given string containing both upper and lower case alphabets. write a python program to count the number of occurences of each
#alphabet(case sensitive) and display the same.
string = input("Enter the string: ")

count = {}

for ch in string:
   if ch.isalpha():
      if ch in count:
         count[ch] += 1
else:
    count[ch] = 1
print("Occurrences of each alphabet:")
for key in sorted(count):
   print(key, ":", count[key])
```

```python
#6.Program to count number of unique words in agiven sentence using sets.
sentence = input("Enter a sentence: ")
words = sentence.split()
unique_words = set(words)
count = len(unique_words)
print("Number of unique words:", count)
```

```python
#7.Create 2 sets s1 and s2 of n fruits each by taking input from user and find: a) Fruits which are in both sets s1 and s2 b) Fruits only in s1 but
#not in s2 c) Count of all fruits from s1 and s2
n = int(input("Enter number of fruits in each set: "))
s1 = set()
s2 = set()
print("Enter fruits for set s1:")
for i in range(n):
    s1.add(input())
print("Enter fruits for set s2:")
for i in range(n):
    s2.add(input())
    print("Fruits in both s1 and s2:", s1 & s2)
    print("Fruits only in s1:", s1 - s2)
    print("Count of all fruits:", len(s1 | s2))
```
